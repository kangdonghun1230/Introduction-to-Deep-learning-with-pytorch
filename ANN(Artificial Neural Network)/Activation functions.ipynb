그래프를 그리기 위한 도구를 추가

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sigmoid function and gradient descent
시그모이드 함수를 그리고 gradient descent에 대해서 알아본다.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# -5.0 ~ 5.0을 0.1 단위로 끊어서 list에 저장해라
x = np.arange(-5.0, 5.0, 0.1)
# sigmoid 함수의 입력으로 x를 넣었을 때 output
y = sigmoid(x)

plt.plot(x, y)
plt.plot([0, 0], [1.0, 0.0], ':') # 가운데 점선을 추가한다.
plt.title('Sigmoid function') # title : Sigmoid function

plt.show()
```

## Hyperbolic tangent function
하이퍼볼릭탄젠트 함수는 입력값을 -1 ~ 1 범위의 값으로 변환합니다. 그래프를 그리면 아래와 같습니다.

```python
x_tanh = np.arange(-5.0, 5.0, 0.1)
y_tanh = np.tanh(x_tanh)

plt.plot(x_tanh, y_tanh)
plt.plot([0, 0], [1.0, -1.0], ':')
plt.axhline(y=0, color='orange', linestyle='--')
plt.title('Tanh Function')
plt.show()
```

## ReLU
인공 신경망에서 가장 많이 사용하는 함수입니다. 수식은 f(x) = max(0, x)로 아주 간단합니다.

```python
def relu(x):
    return np.maximum(0, x)

x_relu = np.arange(-5.0, 5.0, 0.1)
y_relu = relu(x)

plt.plot(x_relu, y_relu)
plt.plot([0, 0], [5.0, 0], ':')
plt.title('Relu Function')
plt.show()
```

## Leaky ReLU
죽은 렐루를 보완하기 위해 ReLU의 변형 함수들이 등장하기 시작했고, LeakyReLU는 그 예시 중 하나입니다.<br>
수식은 f(x) = max(ax, x)로 아주 간단합니다.

```python
a = 0.1

def leaky_relu(x):
    return np.maximum(a*x, x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.plot([0,0],[5.0,0.0], ':')
plt.title('Leaky ReLU Function')
plt.show()
```

## Softmax function
hidden layer에서 ReLU 또는 변형 ReLU를 사용하는것이 일반적이지만, 그렇다고 앞서 배운 시그모이드 함수나 소프트맥스 함수를 사용하지 않는 것은 아닙니다. 분류 문제를 해결하기 위해 로지스틱 회귀나 소프트맥스 회귀를 출력층에 적용하여 사용합니다.

```python
x = np.arange(-5.0, 5.0, 0.1) # -5.0부터 5.0까지 0.1 간격 생성
y = np.exp(x) / np.sum(np.exp(x))

plt.plot(x, y)
plt.title('Softmax Function')
plt.show()
```
