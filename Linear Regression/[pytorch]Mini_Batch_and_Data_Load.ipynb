```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import TensorDataset # 텐서 데이터셋
from torch.utils.data import DataLoader # data loader
```

TensorDataset은 기본적으로 텐서를 입력으로 받습니다. 텐서 형태로 데이터를 정의합니다.

```python
x_train = torch.FloatTensor([[73, 80, 75],
                            [93, 88, 93],
                            [89, 91, 90],
                            [96, 98, 100],
                            [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

```python
dataset = TensorDataset(x_train, y_train)
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=True)
```

```python
model = nn.Linear(3, 1)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5)
```

```python
nb_epochs = 50
for epoch in range(nb_epochs + 1):
    for batch_idx, samples in enumerate(dataloader):
        #print(batch_idx)
        #print(samples)
        x_train, y_train = samples
        
        hypothesis = model(x_train)
        
        cost = F.mse_loss(hypothesis, y_train)
        
        optimizer.zero_grad()
        cost.backward()
        optimizer.step()
        
        print('Epoch {:4d}/{} Batch {}/{} Cost: {:.6f}'.format(
        epoch, nb_epochs, batch_idx+1, len(dataloader),
        cost.item()
        ))
```

```python
# 임의의 입력 [73, 80, 75]를 선언
new_input = torch.FloatTensor([[73, 80, 75]])
# 임의의 입력값에 대한 모델의 예측값을 구한다
pred_y = model(new_input)

print('훈련 후 모델의 예측값 : ', pred_y)
```
